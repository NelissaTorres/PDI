```python
# Paso 1: Importar librerías necesarias
import cv2  #procesamiento de imagenes
import numpy as np  #cálculos numéricos
import matplotlib.pyplot as plt  #matemáticas avanzadas
import urllib.request  #descargar archivos desde Internet

# Paso 2: Descargar y cargar la imagen
url_imagen = "https://raw.githubusercontent.com/dilp90/PDI/main/Imagenes/ave.jpg"
ruta_imagen = "ave.jpg"  #nombre con el que se guardará la imagen localment
urllib.request.urlretrieve(url_imagen, ruta_imagen)

# Paso 3: Cargar la imagen en escala de grises
imagen = cv2.imread(ruta_imagen, cv2.IMREAD_GRAYSCALE)  #convierte la imagen a blanco y negro

# Paso 4: Calcular histpgramas
# Método 1: usando cv2.calcHist()
hist_cv2 = cv2.calcHist([imagen], [0], None, [256], [0, 256])
#parámetros: imagen, canal 0 (gris), sin máscara, 256 bins, rango de 0 a 256

# Método 2: usando plt.hist()
pixeles = imagen.ravel()  #aplanar imagen a 1D para plt.hist()


# Paso 5: Mostrar resultados
plt.figure(figsize=(14, 5)) #crea una figura de 3 gráficos horizontale

# Imagen Original
plt.subplot(1, 3, 1)   #primera columna de 3
plt.imshow(imagen, cmap="gray")   #mostrar la imagen en escala de grises
plt.title("Imagen Original")   #título del gráfico
plt.axis("off")   #quitar ejes

# Histograma con cv2.calcHist()
plt.subplot(1, 3, 2)   #segunda columna de 3
plt.plot(hist_cv2, color='black') #dibujar histograma como línea
plt.title("Histograma con cv2.calcHist()") #título del gráfico
plt.xlabel("Intensidad")   #etiqueta del eje X
plt.ylabel("Frecuencia")   #etiqueta del eje Y
plt.grid(True)             #Mostrar cuadrícula

# Histograma con plt.hist()
plt.subplot(1, 3, 3)   #tercera columna de 3
plt.hist(pixeles, bins=256, range=[0, 256], color='gray') #dibujar histograma de barras
plt.title("Histograma con plt.hist()")   #título del gráfico
plt.xlabel("Intensidad")   #etiqueta del eje X
plt.ylabel("Frecuencia")   #etiqueta del eje Y
plt.grid(True)             #Mostrar cuadrícula


plt.tight_layout() #ajustar los espacios entre los gráficos para evitar que se encimen
plt.show()  #mostrar


print("ANÁLISIS DE RESULTADOS:")
print("El histograma generado a partir de la imagen en escala de grises nos permite analizar la distribución tonal de los píxeles, es decir, ")
print("cómo se reparten los niveles de intensidad entre el negro (0) y el blanco (255). En este caso, se utilizaron dos métodos y ambos mostraron resultados consistentes.")

print("- Zonas oscuras predominantes")
print("- Brillo bajo")
print("- Distribución no uniforme")
```
